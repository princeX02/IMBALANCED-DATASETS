```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("salary.csv")
```

```python
data.head(3)
```

```python
age_mean=data["Age"].mean()
data["Age"]=data["Age"].fillna(age_mean)

year_mean=data["Years of Experience"].mean()
data["Years of Experience"]=data["Years of Experience"].fillna(year_mean)

salary_mean=data["Salary"].mean()
data["Salary"]=data["Salary"].fillna(salary_mean)
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x="Age",y="Years of Experience",hue="Salary",data=data)
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
dr=DecisionTreeRegressor()
dr.fit(x_train,y_train)
```

```python
dr.score(x_test,y_test)
```

```python
dr.score(x_train,y_train)
```

```python

plot_tree(dr)
plt.show()
```

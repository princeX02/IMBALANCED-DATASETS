```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
data=pd.read_csv("Placement.csv")
```

```python
data.head(4)
```

```python
data.drop(columns="Student_ID",inplace=True)
```

```python
data.head(3)
```

```python
sns.kdeplot(data=data["IQ"])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="CGPA",y="IQ",data=data,hue="Placement")
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data["Placement"]
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
```

```python
gub=GaussianNB()
gub.fit(x_train,y_train)
```

```python
gub.score(x_train,y_train)*100,gub.score(x_test,y_test)*100
```

```python
gub.predict([[5.9,106]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gub)
plt.show()
```

```python
ber=BernoulliNB()
ber.fit(x_train,y_train)
```

```python
ber.score(x_train,y_train)*100,ber.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=ber)
plt.show()
```

```python
mub=MultinomialNB()
mub.fit(x_train,y_train)
```

```python
mub.score(x_train,y_train)*100,mub.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mub)
plt.show()
```
